```python
_STATUS_ACTIVE = 10001
```

```python
query = "SELECT device.iddevice, device.name, coretype.manuf, coretype.model, \
         location.name as location, status.status as pingstatus, device.lastpingtimestamp as timestamp \
         FROM sensordb_device device \
         JOIN sensordb_coretype coretype ON device.coretype_id=coretype.idcoretype \
         JOIN sensordb_devicestatus status ON device.lastpingstatus=status.iddevicestatus \
         JOIN sensordb_location location ON device.location_id=location.idlocation \
         WHERE device.status_id={0} ORDER BY coretype.model asc".format(_STATUS_ACTIVE)

print(query)

print("Today is {0}".format(datetime.now()))
df1 = psql.read_sql(query, conn)
nRows = df1.size
if nRows > 0:
    print("The query returned the following:") 
    print(df1.head(10000))
    #for index, row in df1.iterrows():
    #  print(row['iddevice'], row['name'], row['manuf'], row['model'], row['location'])
else:
    print("The query returned empty.")
```

*Ping History

```python
query = "SELECT ping.timestamp, ping.devicestatus_id, status.status FROM sensordb_ping ping \
         JOIN sensordb_device device ON device.iddevice=ping.device_id \
         JOIN sensordb_devicestatus status ON ping.devicestatus_id=status.iddevicestatus \
         WHERE ping.device_id={0}".format(_DEVICE_ID)

#print(query)

print("Today is {0}".format(datetime.now()))
df1 = psql.read_sql(query, conn)
nRows = df1.size
if nRows > 0:
    print("The query returned the following:")
    print(df1.head(10000))
else:
    print("The query returned empty.")
```

```python
query = "SELECT device.name, event.timestamp, topic.description, event.doc, sensor.units FROM sensordb_event event \
         JOIN sensordb_device device ON device.iddevice=event.device_id \
         JOIN sensordb_publishtopic topic ON event.publishtopic_id=topic.idpublishtopic \
         JOIN sensordb_sensortype sensor ON event.sensortype_id=sensor.idsensortype \
         WHERE event.device_id={0} AND event.timestamp>(NOW() - INTERVAL '{1} hours') \
         ORDER BY event.timestamp desc".format(_DEVICE_ID, _HOURS)

#print(query)

print("Today is {0}".format(datetime.now()))
df1 = psql.read_sql(query, conn)
nRows = df1.size
if nRows > 0:
    print("The query returned the following:")
    print(df1.head(10000))
else:
    print("The query returned empty.")
```
