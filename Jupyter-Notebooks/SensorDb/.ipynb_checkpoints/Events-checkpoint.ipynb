```python
import psycopg2 as pg
import pandas.io.sql as psql
from datetime import datetime
conn_str = "host={0} port={1} dbname={2} user={3} password={4}" \
.format("localhost", 5432, "sensordb", "postgres", "w0lfpack")

try:
    conn = pg.connect(conn_str)
    print("Welcome to Jupyter Notebook.  You are connected to the SensorNet database!")
except pg.OperationalError:
    print("You are not connected to the database.")
```

```python
_HOURS = 20
_EVENT_TYPE = 16
_STATUS_ACTIVE = 10001
```

```python
query = "SELECT device.iddevice FROM sensordb_device device WHERE device.status_id={0}".format(_STATUS_ACTIVE)

#print(query)

print("Today is {0}".format(datetime.now()))
df1 = psql.read_sql(query, conn)
nRows = df1.size
if nRows > 0:
    print("The query returned the following:") 
    for index, row in df1.iterrows():
      _DEVICE_ID = row[0]
      query = "SELECT device.name, event.timestamp, topic.description, event.doc, sensor.units FROM sensordb_event event \
         JOIN sensordb_device device ON device.iddevice=event.device_id \
         JOIN sensordb_publishtopic topic ON event.publishtopic_id=topic.idpublishtopic \
         JOIN sensordb_sensortype sensor ON event.sensortype_id=sensor.idsensortype \
         WHERE device.iddevice={0} AND event.timestamp>(NOW() - INTERVAL '{0} hours') \
         ORDER BY event.timestamp desc".format(_DEVICE_ID, _HOURS)
      df1 = psql.read_sql(query, conn)
      nRows = df1.size
      if nRows > 0:
         print("The query returned the following:")
         print(df1.head(10000))
         #for index, row in df1.iterrows():
         #  print(row)
      else:
        print("The query returned empty.")      
else:
    print("The query returned empty.")


```
